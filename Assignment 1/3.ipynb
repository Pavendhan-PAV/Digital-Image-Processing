```python
import cv2
from matplotlib import pyplot as plt 
```

```python
#COLOR IMAGE - ORIGINAL
img = cv2.imread("Lenna.png")
print(img.shape)
#cv2.imshow('ORIGINAL',img)
#cv2.waitKey(4000)
#cv2.destroyAllWindows()
plt.title('ORIGINAL')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#GRAY IMAGE
gray_img = cv2.imread("Lenna.png",0)
print(gray_img.shape)
#cv2.imshow('GRAY',gray_img)
#cv2.waitKey(4000)
#cv2.destroyAllWindows()
plt.title('Gray-Scale')
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#SALT & PEPPER NOISE
from skimage.util import random_noise
import numpy as np

def salt_pepper():
    sp_img = random_noise(gray_img,mode = 's&p',amount = 0.15)
    sp_img = np.array(255*sp_img,dtype=np.uint8)
    return sp_img

def find_avg_sp(n):
    li = []
    for i in range(n):
        li.append(salt_pepper())
    avg_img = np.mean(li,axis = 0)
    avg_img = avg_img.astype(np.uint8)
    return avg_img

img = []
img.append(find_avg_sp(5))
img.append(find_avg_sp(10))
img.append(find_avg_sp(15))
img.append(find_avg_sp(20))
img.append(find_avg_sp(25))
img.append(find_avg_sp(30))

for i in range(6):
    plt.subplot(2,3,i+1), plt.imshow(img[i],'gray')
```

```python
#GAUSSIAN NOISEnoise2_img

def gaussian():
    gaus_img = random_noise(gray_img,mode='gaussian',mean = 0,var = 0.01)
    gaus_img = np.array(255*gaus_img, dtype = np.uint8)
    return gaus_img

def find_avg_g(n):
    li=[]
    for i in range(n):
        li.append(gaussian())
    avg_img = np.mean(li,axis = 0)
    avg_img = avg_img.astype(np.uint8)
    return avg_img;

img = []
img.append(find_avg_g(5))
img.append(find_avg_g(10))
img.append(find_avg_g(15))
img.append(find_avg_g(20))
img.append(find_avg_g(25))
img.append(find_avg_g(30))

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(img[i],'gray')
```

```python
#SPECKLE NOISE

def speckle():
    row,col = gray_img.shape
    gauss = np.random.randn(row,col)
    gauss = gauss.reshape(row,col)        
    noisy = gray_img + gray_img * gauss
    return noisy

def find_avg_sk(n):
    li=[]
    for i in range(n):
        li.append(speckle())
    avg_img = np.mean(li,axis=0)
    avg_img = avg_img.astype(np.uint8)
    return avg_img

img = []
img.append(find_avg_sk(5))
img.append(find_avg_sk(10))
img.append(find_avg_sk(15))
img.append(find_avg_sk(20))
img.append(find_avg_sk(25))
img.append(find_avg_sk(30))

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(img[i],'gray')
```
